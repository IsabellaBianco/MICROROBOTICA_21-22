```python
#Esercizio 2

#Creo i due dizionari
frequenze = {} #carattere:frequenza
adiacenze = {}

CARATTERI_DA_RIMUOVERE = ["\n", "\r", "\t"]

#Apro il file .txt
file = open("./Isabella Bianco - J_Verne_girodelmondoin80giorni.txt", "r")

for riga in file.readlines():
    #Verifica se il carattere è da escludere
    riga = ''.join(i for i in riga if not i in CARATTERI_DA_RIMUOVERE)
    for carattere in riga:
        #Aggiorna il dizionario delle frequenze
        if(carattere in frequenze.keys()):
            frequenze[carattere] +=1
        else:
            frequenze[carattere] = 1
            adiacenze[carattere] = []
        if riga.index(carattere)+1 <= len(riga)-1:
            if adiacenze[riga[riga.index(carattere)+1] not in adiacenze[carattere]:
                         adiacenze[carattere].append(riga[riga.index(carattere)+1])
                         
print(f"Dizionario di frequenza {frequenze}")
print(f"Lunghezza {len(frequenze)}")
#print(f"\n\n\n\n\n\n\n\n\n\n Dizionario di adiacenze {adiacenze}")
    
file.close()
```
