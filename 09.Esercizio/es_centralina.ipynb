### ES 2 - Centralina scolastica

Usiamo Numpy per creare dei dati simulati riferiti ad una situazione reale: 
in certi casi i dati simulati sono una buona approssimazione della realtà. 
Nel parco vicino alla scuola è installata una centralina per la misurazione del livello di polveri sottili 
PM10 nell'aria. Talvolta la centralina va in uno stato di errore per cui la misurazione delle PM10 fallisce 
ed il dato corrispondente riporta il valore 9999.9 all'interno dell'array delle misurazioni dati. 
Creare un array di 1000 campioni che generino dei valori secondo una distribuzione normale con valore medio 19 
e varianza 5 e supporre che la centralina sia andata in errore 20 volte. La distribuzione gaussiana è una buona
approssimazione dei dati reali della centralina?



```python
import numpy as np
```

```python
misurazioni = np.array(np.random.normal(loc=19, scale=np.sqrt(5), size=1000))
misurazioni[(np.random.choice(range(1000),20,replace=False))] = 9999.99
print(misurazioni)
```

### ES 1 
Creare una funzione che trovi gli elementi comuni a due ndarray monodimensionali di lunghezza diversa.

```python
arr_2 = np.array([1,2,3,4,5,6])
arr_1 = np.array([4,3,1,8,5])

arr_somma =np.array([arr_1[a] for a in range(arr_1.size) for b in range(arr_2.size) if arr_1[a] == arr_2[b]])
print(arr_somma)
```
